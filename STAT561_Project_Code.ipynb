```python
def create_dnn_model(input_shape):
    model = Sequential([
        Dense(16, activation='relu', input_shape=input_shape),
        Dense(1, activation='sigmoid')  # Binary classification
    ])

    model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['AUC'])

    return model
```
